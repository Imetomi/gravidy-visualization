```python
import time
from IPython.display import clear_output
```

```python
for i in range(1000):
    clear_output(wait=True)
    data = open('../output/experiment1/experiment.out.snapshot_' + str(i).zfill(5), 'r')
    data.readline()
    particles = np.array([np.fromstring(line, dtype=float, sep=" ") for line in data.readlines()])
    x = particles[:,2]
    y = particles[:,3]
    colors = particles[:,1]
    area = np.log10(particles[:,1]*130) * 200
    fig = plt.figure(figsize=(10, 10))
    ax = plt.gca()
    ax.set_xlim([-1, 1])
    ax.set_ylim([-1, 1])
    ax.set_facecolor('black')
    plt.scatter(x, y, s=area, c=colors, cmap='Oranges')
    plt.show()
```

```python
for i in range(20):
    clear_output(wait=True)
    data = open('../output/experiment1/experiment.out.snapshot_' + str(i).zfill(5), 'r')
    data.readline()
    particles = np.array([np.fromstring(line, dtype=float, sep=" ") for line in data.readlines()])
    x = particles[:,2]
    y = particles[:,3]
    z = particles[:,4]
    colors = particles[:,1]
    area = np.log10(particles[:,1]*9000) * 100
    
    fig = plt.figure(figsize=(10, 10))
    ax = fig.add_subplot(111, projection='3d')
    ax.set_facecolor('black')
    ax.set_xlim([-1, 1])
    ax.set_ylim([-1, 1])
    ax.set_zlim([-1, 1])
    ax.scatter(x, y, z, s=area, c=colors, cmap='Oranges')
    plt.axis('off')
    plt.show()
```
